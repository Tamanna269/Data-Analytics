```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
sns.countplot(data=penguins, x='species')
```

```python
sns.countplot(data=penguins, x ='species', hue='island')
```

```python
sns.countplot(data=penguins, x ='island', hue='species')
```

```python
sns.countplot(data=penguins, y ='island', hue='species', fill=False)
```

```python
sns.histplot(data=penguins,
             x='body_mass_g',
             kde=True,
             bins=10)
```

```python
sns.swarmplot(
    data=penguins,
    x='island',
    y='body_mass_g',
    hue='species'
)
```

```python
sns.relplot(
    data=penguins,
    x='flipper_length_mm',
    y='body_mass_g',
    hue='species',
    col='species'
)
```

```python
sns.relplot(
    data=penguins,
    x='flipper_length_mm',
    y='bill_depth_mm',
    hue='species',
    col='species',
    row='island'
)
```

```python
sns.relplot(
    data=penguins,
    x='flipper_length_mm',
    y='bill_depth_mm',
    hue='species',
    col='sex'
)
```

```python
px.scatter_3d(
    data_frame=penguins,
    x = 'flipper_length_mm',
    y= 'bill_depth_mm',
    z= 'body_mass_g',
    color='species'

)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue ='sex'
)
```

```python
sns.catplot(
    data=tips,
    x='day',
    y='total_bill',
    hue ='smoker',
    kind = 'swarm'
)
```

```python
sns.catplot(
    data = tips ,
    x = 'day',
    #y = 'total_bill',
    hue = 'smoker',
    kind = 'count'

)
```

```python
sns.catplot(
    data = tips,
    x='time',
    y='tip',
    hue='sex',
    kind='box'
    )
```

```python
# sunburst
px.sunburst(
    data_frame=tips,
    path=['day','time','sex','smoker'],
    values='tip'
)
```

```python

px.treemap(
    data_frame=tips,
    path=['day','time','sex','smoker'],
    values='tip'
)
```
